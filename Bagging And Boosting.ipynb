```python
# seperating Numerical And Categirical Columns
categorical_columns = df2.loc[:,df2.dtypes == object]
numerical_columns = df2.loc[:,((df2.dtypes =='float64')|(df2.dtypes =='int64'))]
```

```python
#Label Encoding for Categorical type Data
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for x in categorical_columns:
    categorical_columns[x]= le.fit_transform(categorical_columns[x])
```

```python
#Concatination of numerical and categorical data columns
df3= pd.concat([categorical_columns,numerical_columns], axis='columns')
```

# Spliting And Training the model

```python
# indipendent variable Data frame (X)
X = df3.drop(['Loan_Status'], axis='columns')
# Dependent variable Data frame
y = df3.Loan_Status
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
# training the model
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_params = {
    'svm':{'model':SVC(gamma='auto'), 'params':{'kernel':['rbf','linear']} },
    
    'RandomForestClassifier':{'model': RandomForestClassifier(),'params':{'n_estimators':[1,10,20]}},
    
    'LogisticRegression':{'model':LogisticRegression(solver='liblinear',multi_class='auto'),'params':{'C':[1,10,30]}},
        
    'KNeighborsClassifier':{'model': KNeighborsClassifier(),'params': {}},
    
    'Decision_tree':{'model': DecisionTreeClassifier(),'params': {'criterion': ['gini','entropy']}}
    
} 
```

```python
from sklearn.model_selection import GridSearchCV
scores = []

for model_name, mp in model_params.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv=3)
    clf.fit(X_train,y_train)
    scores.append({'model': model_name,'best_score': clf.best_score_})
```

```python
df_scores = pd.DataFrame(scores,columns=['model','best_score'])
df_scores
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import confusion_matrix as cm
from sklearn.ensemble import AdaBoostClassifier
```

```python
# Using Bagging and Boosting technique to reduce Bias and Variance 
# Bagging
bag_clf = BaggingClassifier(DecisionTreeClassifier(random_state=29), n_estimators=500,max_samples=100,
                            bootstrap=True, n_jobs=-1, random_state=29)
bag_clf.fit(X_train, y_train)

# Boosting
ada_clf = AdaBoostClassifier(DecisionTreeClassifier(max_depth=1), n_estimators=200,learning_rate=0.5, random_state=50)

ada_clf.fit(X_train, y_train)
```

```python
bag_clf.score(X_test,y_test)
```
